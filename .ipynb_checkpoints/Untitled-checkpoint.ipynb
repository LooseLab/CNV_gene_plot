```python
import os
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import click

from cnv_from_bam import iterate_bam_file 

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 3))
chosen_chr = "chr3"
bam_path = Path("/home/thomas/sort_ds1305_Intraop0002_2.htom.bam")
CNS2_gene_bed_path = "rCNS2_panel_name_uniq.bed"
AML_gene_bed_path = "AML_panel_name_uniq.bed"
CNS2_gene_bed_df = pd.read_csv(CNS2_gene_bed_path, sep='\t', header=None)
AML_gene_bed_df = pd.read_csv(AML_gene_bed_path, sep='\t', header=None)

bed_array = CNS2_gene_bed_df.to_numpy()


result = iterate_bam_file(bam_path, _threads=4, mapq_filter=60)
bin_width = result.bin_width

# Iterate over the BAM file and calculate CNV values for each bin.
for contig, cnv in result.cnv.items():
    if contig == chosen_chr:
        ax.scatter(x=np.arange(len(cnv)) * bin_width, y=cnv, s=0.1)

for array in bed_array:
    if array[0] == chosen_chr:
        start = array[1]
        end = array[2]
        label = str(array[3])
        ax.axvspan(start, end, alpha=0.5, color='red')
        ax.text((start + end) / 2, 8, label, rotation=90, verticalalignment='bottom', fontsize=8, color='red')
        print(array)

ax.set_ylim((0, 8))

plt.show()

```
